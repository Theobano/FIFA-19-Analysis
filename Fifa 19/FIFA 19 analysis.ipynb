### Dropping some columns

```python
data.drop(['ID','Photo', 'Flag','Club Logo','International Reputation','Body Type', 'Real Face'], axis=1,inplace=True)
```

### Grouping the data by club

```python
clubs=data.groupby('Club')
```

### The club with the highest average overall rating

```python
club_name=[]
club_avgrat=[]
for i, j in clubs:
    club_name.append(i)
    club_avgrat.append(j['Overall'].mean())
club_rating = pd.DataFrame({'club':club_name, 'average_rating':club_avgrat})

club_rating[club_rating['club']=='Juventus']
```

Turns out Juventus is the club with the highest average overall rating of their players.

The rest of the ranking is below

```python
club_rating.sort_values(by='average_rating',axis=0, ascending=False)
```

### The fastest players

```python
data.sort_values(by='Acceleration', axis=0, ascending=False)[['Name', 'Age','Nationality','Overall', 'Potential', 'Club',
       'Preferred Foot','Work Rate','Position','Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'ShotPower',
       'Jumping', 'Stamina', 'LongShots', 'Positioning', 'Vision', 'Composure']].head()
```

The player with highest acceleration is the spanish Adama who plays for Wolverhanpton Wanderers. He is followed by Juventus' Douglas Costa.

```python
data.sort_values(by='SprintSpeed', axis=0, ascending=False)[['Name', 'Age','Nationality','Overall', 'Potential', 'Club',
       'Preferred Foot','Work Rate','Position','Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'ShotPower',
       'Jumping', 'Stamina', 'LongShots', 'Positioning', 'Vision', 'Composure']].head()
```

Adama is also the player with the highest speed, followed by L. Sane.

### Grouping by position

```python
position= data.groupby('Position')
```

```python
df_pos = []
poss = []
for pos, group in position:
    top_pos = position.get_group(pos)[:2]
    poss.append(pos)
    df_pos.append(top_pos)
    
pd.concat(df_pos, keys=poss).drop(columns=["Unnamed: 0"])
```

The above cell shows the top two players playing each position.

### Grouping by Nationality

```python
countries = data.groupby('Nationality')
```

```python
country_name=[]
country_avgrat=[]
for i, j in countries:
    country_name.append(i)
    country_avgrat.append(j['Overall'].mean())
country_rating = pd.DataFrame({'country':country_name, 'average_rating':country_avgrat})
```

```python
country_rating.sort_values(by='average_rating',axis=0, ascending=False)
```

Turns out the UAE has the highest average rating of players. This includes all players whose nationality is the UAE and not neccessarily playing for the national team.

### Correlation between attributes

```python
data.drop(columns=["Unnamed: 0"]).corr()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(data=data.drop(columns=["Unnamed: 0"]).corr(), vmin= -1, vmax = 1, annot= False, xticklabels=True, yticklabels= True)
sns.set_context("notebook")
plt.tight_layout()
```

The heatmap shows the correlation between the player attributes.
